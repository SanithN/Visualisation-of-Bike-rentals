Listing the season that fall on respective months


```python
ex3_df = hour_df.loc[hour_df.season == 1]
ex4_df = hour_df.loc[hour_df.season == 2]
ex5_df = hour_df.loc[hour_df.season == 3]
ex6_df = hour_df.loc[hour_df.season == 4]
print('Season 1 occurs from December to next years March')
print(ex3_df.groupby(['yr']).mnth.value_counts())
print('Season 2 occurs from midway March to June ')
print('',ex4_df.groupby(['yr']).mnth.value_counts())
print('Season 3 occurs from midway June to September ')
print('',ex5_df.groupby(['yr']).mnth.value_counts())
print('Season 4 occurs from midway September to December ')
print('',ex6_df.groupby(['yr']).mnth.value_counts())
#sns.countplot(x="mnth",hue="yr", data=ex3_df)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(13, 6))
#fig.suptitle('ANALYZING SEASONS AND MONTHS wrt RENTALS')
#fig,ax = plt.subplots()
ax1 = sns.countplot(x="mnth",hue="yr", data=ex3_df,ax=ax1)
ax1.set(title='count of records for season 1')
ax2 = sns.countplot(x="mnth",hue="yr", data=ex4_df,ax=ax2)
ax2.set(title='Count of records for season 2')
ax3 = sns.countplot(x="mnth",hue="yr", data=ex5_df,ax=ax3)
ax3.set(title='Count of records for season 3')
ax4 = sns.countplot(x="mnth",hue="yr", data=ex6_df,ax=ax4)
ax4.set(title='Count of records for season 4')
```
